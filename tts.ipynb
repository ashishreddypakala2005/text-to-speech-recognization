```python
!pip install gTTS

from gtts import gTTS
from IPython.display import Audio, display, clear_output

def text_to_speech():
    try:
        text = input("Enter text and press Enter: ")

        if text.strip() == "":
            print("Please enter some text!")
            return

        tts = gTTS(text=text, lang='en')
        tts.save("output.mp3")

        clear_output(wait=True)  # clears input line nicely
        display(Audio("output.mp3"))

    except KeyboardInterrupt:
        print("\nInput interrupted. Please run again.")

# Run the function
text_to_speech()
```
